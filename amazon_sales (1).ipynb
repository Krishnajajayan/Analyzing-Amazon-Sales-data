```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('/content/drive/MyDrive/internship_datasets/Amazon_Sales_data.csv')
df.head()
```

```python
data_f=df[['Total Profit','Total Cost','Units Sold']]
```

```python
df.info()
```

```python
total_profit = df['Total Profit'].sum()
print('Total Profit:', total_profit)
total_revenue = df['Total Revenue'].sum()
print('Total Revenue:',total_revenue)
```

```python
percentage = []
for i in range(df.shape[0]):
    pct = (df.loc[i, 'Total Profit'] / total_revenue) * 100
    percentage.append(round(pct, 2))
print(percentage)
df['percentage']=percentage
display(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
colors_list = ['Red', 'Green', 'Orange', 'Purple', 'blue', 'yellow', 'pink']
graph = plt.bar(df['Region'], df['Total Profit'], color=colors_list)
plt.title('Percentage across of all formats')

i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()

    plt.text(x+width/2,
             y+height*1.01,
             str(df.percentage[i])+'%',
             ha='center',
             weight='bold')
    i += 1
plt.show()
```

```python
# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df.dtypes
```

```python
# Extract year and month from 'Order Date'
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# Calculate month-wise sales trend
month_wise_sales = df.groupby('Month')['Total Revenue'].sum().reset_index()
month_wise_sales.columns = ['Month', 'Total Revenue']
```

```python
# Calculate year-wise sales trend
year_wise_sales = df.groupby('Year')['Total Revenue'].sum().reset_index()
year_wise_sales.columns = ['Year', 'Total Revenue']
```

```python
# Calculate yearly month-wise sales trend
year_month_wise_sales = df.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()
year_month_wise_sales.columns = ['Year', 'Month', 'Total Revenue']
```

```python
import plotly.express as px
```

```python
# Plot month-wise sales trend
fig_month_wise = px.line(month_wise_sales, x='Month', y='Total Revenue', title='Month-wise Sales Trend')
fig_month_wise.show()
```

```python
# Plot year-wise sales trend
fig_year_wise = px.line(year_wise_sales, x='Year', y='Total Revenue', title='Year-wise Sales Trend')
fig_year_wise.show()
```

```python
# Plot yearly month-wise sales trend
fig_yearly_month_wise = px.line(year_month_wise_sales, x='Month', y='Total Revenue', color='Year', title='Yearly Month-wise Sales Trend')
fig_yearly_month_wise.show()
```

```python
# Total Revenue
total_revenue = df['Total Revenue'].sum()

# Total Profit
total_profit = df['Total Profit'].sum()

# Average Revenue per Order
avg_revenue_per_order = df['Total Revenue'].mean()

# Average Profit per Order
avg_profit_per_order = df['Total Profit'].mean()

# Top Countries by Revenue
top_countries_by_revenue = df.groupby('Country')['Total Revenue'].sum().reset_index().sort_values(by='Total Revenue', ascending=False)

# Top Products by Revenue
top_products_by_revenue = df.groupby('Item Type')['Total Revenue'].sum().reset_index().sort_values(by='Total Revenue', ascending=False)

# Monthly Revenue Trend
monthly_revenue_trend = df.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()

# Monthly Profit Trend
monthly_profit_trend = df.groupby(['Year', 'Month'])['Total Profit'].sum().reset_index()

# Yearly Revenue Trend
yearly_revenue_trend = df.groupby('Year')['Total Revenue'].sum().reset_index()

# Yearly Profit Trend
yearly_profit_trend = df.groupby('Year')['Total Profit'].sum().reset_index()

print(f"Total Revenue: {total_revenue}")
print(f"Total Profit: {total_profit}")
print(f"Average Revenue per Order: {avg_revenue_per_order}")
print(f"Average Profit per Order: {avg_profit_per_order}")

print("\nTop Countries by Revenue:")
print(top_countries_by_revenue.head(10))

print("\nTop Products by Revenue:")
print(top_products_by_revenue.head(10))

print("\nMonthly Revenue Trend:")
print(monthly_revenue_trend)

print("\nMonthly Profit Trend:")
print(monthly_profit_trend)

print("\nYearly Revenue Trend:")
print(yearly_revenue_trend)

print("\nYearly Profit Trend:")
print(yearly_profit_trend)
```

```python
# Top 10 Countries by Revenue
fig_top_countries = px.bar(top_countries_by_revenue.head(10), x='Country', y='Total Revenue', title='Top 10 Countries by Revenue')
fig_top_countries.show()

# Top 10 Products by Revenue
fig_top_products = px.bar(top_products_by_revenue.head(10), x='Item Type', y='Total Revenue', title='Top 10 Products by Revenue')
fig_top_products.show()

# Monthly Revenue Trend
fig_monthly_revenue_trend = px.line(monthly_revenue_trend, x='Month', y='Total Revenue', color='Year', title='Monthly Revenue Trend')
fig_monthly_revenue_trend.show()

# Monthly Profit Trend
fig_monthly_profit_trend = px.line(monthly_profit_trend, x='Month', y='Total Profit', color='Year', title='Monthly Profit Trend')
fig_monthly_profit_trend.show()

# Yearly Revenue Trend
fig_yearly_revenue_trend = px.line(yearly_revenue_trend, x='Year', y='Total Revenue', title='Yearly Revenue Trend')
fig_yearly_revenue_trend.show()

# Yearly Profit Trend
fig_yearly_profit_trend = px.line(yearly_profit_trend, x='Year', y='Total Profit', title='Yearly Profit Trend')
fig_yearly_profit_trend.show()
```
